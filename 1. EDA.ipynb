# Correlation

Let's investigate correlations of features between each other to:
- check for multicollinearity that can affect modeling
- checks if there are any features that bring in more information about target variable

I would avoid calculate correlation coefficients for variables, that have unsuitable distribution such as binary variables for both Pearson and Spearman coefficient and non linear distribution for Pearson.

Pearson correlation of features, without 8 (binary) and 6 (non linear distribution) columns:

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    df\
        .drop(columns = ['8', '6', 'target'])\
        .corr(method = 'pearson'), 
    annot=False, 
    cmap='coolwarm', 
    center=0, 
    vmin=-1, 
    vmax=1, 
    square=True
)
plt.show()
```

Spearman correlation of features, without 8 (binary) column:

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    df\
        .drop(columns = ['8', 'target'])\
        .corr(method = 'spearman'), 
    annot=False, 
    cmap='coolwarm', 
    center=0, 
    vmin=-1, 
    vmax=1, 
    square=True
)
plt.show()
```

Looks like variables are independent, no multicollinearity seen

Correlation with target without '8' column (binary) for both and without '6' only for Spearman:

```python
target_pearson = df\
    .corr(method = 'pearson')\
    .target\
    .to_frame()\
    .rename(columns = {'target':'target_pearson_correlation'})

target_pearson.loc['6'] = np.nan

target_spearman = df\
    .corr(method = 'spearman')\
    .target\
    .to_frame()\
    .rename(columns = {'target':'target_spearman_correlation'})

target_pearson\
    .merge(target_spearman, left_index = True, right_index = True)\
    .drop(index = ['8', 'target'])\
    .reset_index()\
    .rename(columns = {'index':'variable'})\
    .sort_values(by = 'target_pearson_correlation', ascending = False)\
    .style\
        .format({
            'target_pearson_correlation':'{:,.3f}',
            'target_spearman_correlation':'{:,.3f}',
        })\
        .background_gradient(subset=['target_pearson_correlation', 'target_spearman_correlation'])
    
```

```python
# visualization of variables against the target

cols = 4 # columns per row
n_plots = df.shape[1] - 1  # excluding the 'target' column
rows = (n_plots // cols) + (n_plots % cols > 0)

fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 5))
axes = axes.flatten()

for i, column in enumerate(df.columns[:-1]):
    axes[i].scatter(df[column], df['target'], alpha=0.5)
    axes[i].set_title(f'Scatter: {column} vs target')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('target')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Looks like there is no variables except for '6' that have visible correlation with target

variable '6' doesn't have visible correlation in terms of Spearman coefficient, but by it's plot against target looks like it has some relationship:

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(
    data = df,
    x = '6',
    y = 'target',
)
plt.title('Variable 6 against target')
plt.grid()
plt.show()
```

Looks like variable 6 related to target as `target ~ abs(var6)**2`, let's check it

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(
    data = df,
    x = '6',
    y = 'target',
    label = 'target'
)

sns.scatterplot(
    data = df.assign(var6_power2 = lambda df_: np.power(np.abs(df_['6']), 2)),
    x = '6',
    y = 'var6_power2',
    label = 'abs(var6)**2'
)
plt.title('Variable 6 against target and abs(var6)**2')
plt.grid()
plt.legend()
plt.show()


plt.figure(figsize=(8, 8))
sns.scatterplot(
    data = df\
        .assign(var6_power2 = lambda df_: np.power(np.abs(df_['6']), 2))\
        .assign(delta = lambda df_: df_.target - df_.var6_power2),
    x = '6',
    y = 'delta',
)
plt.title('Delta between target and abs(var6)**2')
plt.grid()
plt.show()
```

Well, it's close, but not exactly it, look's like there is something else too: residuals are shifted, all are in [0, 1]. Guess, we know more from modeling. Let's proceed in "2. Model selection" notebook.
